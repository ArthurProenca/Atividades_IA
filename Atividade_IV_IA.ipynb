### Funções que iremos usar durante os métodos

```python
# Compara as labels geradas pelo kmeans com os valores originais da classificação
def errors_by_group(kmeansLabels, originalLabels):
  i, errors = (0, 0)
  for kml in kmeansLabels:
    if kml == originalLabels[i]+1:
      errors = errors + 1
    i = i + 1
  return (errors/kmeansLabels.size) * 100

# Atribui os operadores matemáticos para a base passada (atv 3)
def take_math(df, operation, random):
  numeric_df = 0
  if operation == "mult":
    numeric_df = df.apply(lambda x: x*x)
  if operation == "sum":
    numeric_df = df.apply(lambda x: x+12309812)
  if operation == "sub":
    numeric_df = df.apply(lambda x: x*2)
  if operation == "log":
    numeric_df = df.apply(lambda x: np.log(x, out=np.zeros_like(x), where=(x>0)))
  if operation == "sqrt":
    numeric_df = df.apply(lambda x: np.sqrt(x, out=np.zeros_like(x), where=(x>0)))
  if operation == "am":
    numeric_df = df.apply(lambda x: df.mean())
  if operation == "pm":
    if random:
      weight = [1, 1, 2, 1, 2, 1, 1, 1, 2, 1, 2, 1, 2, 1, 2, 2, 2, 1, 2, 2, 2]
    else:
      weight = [1, 1, 2]
    df['pm'] = np.average(df, weights=weight, axis=1)
    numeric_df = df
  if operation == "sen":
    numeric_df = df.apply(lambda x: (np.sin(x)))
  if operation == "cos":
    numeric_df = df.apply(lambda x: (np.cos(x)))

  (kMeansResultsLabels, kMeansResultsCenters) = kMeans(numeric_df)
  errorPercentage = errors_by_group(kMeansResultsLabels, np.array(fetos['fetal_health']))
  print(f"The errsubor percentage on {operation} are {errorPercentage}%")


# Gera trio aleatório (atividade 2)
def generate_results(size):
  tabular = [[]]
  resultArray = []
  fetos_ = fetos.drop(['K-classes', 'fetal_health'], axis=1)
  for n in list(range(0, size)):
    feature1, feature2, feature3 = (random.randint(0, 20), random.randint(0, 20), random.randint(0, 20))
    fetosAtividade2 = fetos_[[fetos_.columns[feature1], fetos_.columns[feature2], fetos_.columns[feature3]]]
    dadosNaoRotulados2 = np.array(fetosAtividade2)
    (kMeansResultsLabels, kMeansResultsCenters) = kMeans(dadosNaoRotulados2)
    errorPercentage = errors_by_group(kMeansResultsLabels, np.array(fetos['fetal_health']))
    tabular.append([n, errorPercentage, fetos_.columns[feature1],  fetos_.columns[feature2],  fetos_.columns[feature3]])
    resultArray.append(errorPercentage)
  
  mean = 0
  for result in resultArray:
    mean = mean + result
  
  print(tabulate(tabular, headers=['Execution', 'Result', 'Feature1', 'Feature2', 'Feature3'], tablefmt='orgtb1'))
  print(f"\n\nThe arithmetic mean of the {size} executions are: {mean/size}%")
  return pd.DataFrame(dadosNaoRotulados2, columns = [fetos_.columns[feature1], fetos_.columns[feature2], fetos_.columns[feature3]])
```

### Base de dados

```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.cluster import KMeans

# Importa o csv
fetos = pd.read_csv("https://raw.githubusercontent.com/SprigganCS/mushroomsdb/main/fetal_health.csv")

fetos.head(5)
```

### K-Means

```python
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)
def kMeans(dadosNaoRotulados):
  kmeans = KMeans(n_clusters=3, random_state=0)

  # Treinando o modelo
  kmeans.fit(dadosNaoRotulados)

  # Saída de rótulos
  kmeansLabels = kmeans.predict(dadosNaoRotulados)
  centers = kmeans.cluster_centers_
  return (kmeansLabels, centers)
```
